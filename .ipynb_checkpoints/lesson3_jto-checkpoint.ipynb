### Remove dropout
1. Re-create and load our modified VGG model with binary dependent (i.e. dogs v cats)
2. Split the model between the convolutional (conv) layers and the dense layers
3. Pre-calculate the output of the conv layers, so that we don't have to redundently re-calculate them on every epoch
    1. Collect the training/validation data and classes in batches, then predict the outcome with the compiled convolutional model.
    2. Save results for feeding into the fully connected part of the model
4. Create a new model with just the dense layers, and dropout p set to zero
    1. Copy the weights from each layer of the old model and half as as we're removing dropout
    2. Create new architecture same as the old FC architecture but set dropout=0 in dropout layers
    3. set_weights for each layer and compile
5. Train this new model using the output of the conv layers as training data.

```python
path = "data/dogscats//"
model_path = path + 'models/'
```

```python
import vgg16; reload(vgg16)
from vgg16 import Vgg16
vgg = Vgg16()
model = vgg.model
```

```python
model.pop()
#for layer in model.layers: layer.trainable=False
model.add(Dense(2, activation='softmax'))
model.load_weights('data//dogscats//models//finetune4.h5')
```

```python
last_conv_idx = [index for index,layer in enumerate(model.layers)
                if type(layer) is Convolution2D][-1]
conv_layers = model.layers[:last_conv_idx+1]
fc_layers = model.layers[last_conv_idx+1:]
```

```python
conv_model = Sequential(conv_layers)
```

```python
def get_batches(dirname, gen=image.ImageDataGenerator(), shuffle=False, batch_size=4, class_mode='categorical',
                target_size=(224,224)):
    return gen.flow_from_directory(dirname, target_size=target_size,
            class_mode=class_mode, shuffle=shuffle, batch_size=batch_size)

import bcolz
def save_array(fname, arr): c=bcolz.carray(arr, rootdir=fname, mode='w'); c.flush()
def load_array(fname): return bcolz.open(fname)[:]

def onehot(x): return np.array(OneHotEncoder().fit_transform(x.reshape(-1,1)).todense())
```

```python
batch_size = 32
trn_batches = get_batches(path+'train//', batch_size=batch_size)
val_batches = get_batches(path+'valid//', batch_size=batch_size)

val_classes = val_batches.classes
trn_classes = trn_batches.classes
val_labels = onehot(val_classes)
trn_labels = onehot(trn_classes)
```

```python
val_features = conv_model.predict_generator(val_batches, val_batches.nb_sample)
trn_features = conv_model.predict_generator(trn_batches, trn_batches.nb_sample)
```

```python
save_array(model_path+'valid_last_conv_layer_features.bc', val_features)
save_array(model_path+'train_last_conv_layer_features.bc', trn_features)
```

```python
val_features = load_array(model_path+'valid_last_conv_layer_features.bc')
trn_features = load_array(model_path+'train_last_conv_layer_features.bc')
```

```python
def proc_weights(layer):
    return [weight/2 for weight in layer.get_weights()]

weights = [proc_weights(layer) for layer in fc_layers]
```

```python
fc_model = Sequential([
        MaxPooling2D(input_shape=conv_layers[-1].output_shape[1:]),
        Flatten(),
        Dense(4096, activation='relu'),
        Dropout(0.),
        Dense(4096, activation='relu'),
        Dropout(0.),
        Dense(2, activation='softmax')
        ])

for idx in range(len(fc_model.layers)):
    fc_model.layers[idx].set_weights(weights[idx])

opt = RMSprop(lr=0.00001, rho=0.7) #finely tuned so update slowly
fc_model.compile(optimizer=opt, loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
fc_model.fit(trn_features, trn_labels, nb_epoch=1,
             batch_size=batch_size, validation_data=(val_features, val_labels))
```

```python
fc_model.save_weights(model_path+'no_dropout.h5')
```

### Adding *batchnorm* to the FC model

1. Add batchnorm layers into the FC part of the model and set dropout layers' p=0.6
2. Load the FC weights from the dense part of Vgg16BN
3. Create bn_model and set weights
4. Pop final layer, set bn_model layers to untrainable and add 2 node softmax layer
5. Compile and fit

```python
def get_bn_layers(p):
    return [
        MaxPooling2D(input_shape=conv_layers[-1].output_shape[1:]),
        Flatten(),
        Dense(4096, activation='relu'),
        BatchNormalization(),
        Dropout(p),
        Dense(4096, activation='relu'),
        BatchNormalization(),
        Dropout(p),
        Dense(1000, activation='softmax')
        ]

def load_fc_weights_from_vgg16bn(model):
    "Load weights for model from the dense layers of the Vgg16BN model."
    # See imagenet_batchnorm.ipynb for info on how the weights for
    # Vgg16BN can be generated from the standard Vgg16 weights.
    from vgg16bn import Vgg16BN
    vgg16_bn = Vgg16BN()
    _, fc_layers = split_at(vgg16_bn.model, Convolution2D)
    copy_weights(fc_layers, model.layers)
    

def proc_wgts(layer, prev_p, new_p):
    scal = (1-prev_p)/(1-new_p)
    return [o*scal for o in layer.get_weights()]
    
p = 0.6
bn_model = Sequential(get_bn_layers(0.6))
load_fc_weights_from_vgg16bn(bn_model)

for l in bn_model.layers: 
    if type(l)==Dense: l.set_weights(proc_wgts(l, 0.5, 0.6))
```
